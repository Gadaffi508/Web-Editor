```python
from pytrends.request import TrendReq
pytrends = TrendReq()
```

```python
keywords = ["Python","R","C++","Java","HTML"]
pytrends.build_payload(keywords,timeframe="today 5-y")
```

```python
df = pytrends.interest_over_time()
df.head()
```

```python
df.info()
```

```python
del df["isPartial"]
```

```python
df.plot()
```
