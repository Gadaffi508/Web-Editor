```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://bigpara.hurriyet.com.tr/altin/gram-altin-fiyati/1ay/"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
div = soup.find("div",class_="tBody")
div
```

```python
linkler = div.find_all("a")
linkler
```

```python
turler = []
for link in linkler:
    turler.append(link.text)

turler
```

```python
prices = div.find_all("li",class_="cell009")
prices
```

```python
alis = []
stais = []

for i in range (len(prices)):
    if i%4==0:
        alis.append(prices[i].text)
    if i%4==1:
        stais.append(prices[i].text)
```

```python
print("Gold")
for i in range(len(prices)):
    print(turler[i],alis[i],stais[i])
```
