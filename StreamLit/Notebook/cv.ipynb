### Edge 

```python
import cv2
#--> opencv-python paketinin Python'a aktarılan adıdır. import diyerek opencv-python kütüphanesini cv2 yazarak kullanıyoruz
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------

import numpy as np
#--> opencv-python paketinin numpy kütüphanesini import edip np numpy kütüphanesinin sıkça kullanılan bir kısaltmasıdır. 
#--> Numpy, Python programlama dilinde bilimsel hesaplamalar için kullanılan güçlü bir kütüphanedir.
#--> Numpy, çok boyutlu dizilerin (arrays) oluşturulması ve işlenmesi için çeşitli fonksiyonlar ve araçlar sağlar.
#--> Bu, matematiksel işlemler, lineer cebir, rastgele sayı üretimi, Fourier dönüşümleri gibi birçok alanı içerir.
#---------------------------------------------------------------------------------------------------------------------

image = cv2.imread("unnamed.png")
#--> cv2 ile kütüphaneyi çağırıyoruz.
#--> cv2 nin imread fonksiyonunu çağırarak unnamed.png isimli resim dosyasını okutuyoruz.
#--> imread fonksiyonu resim dosyasını okuyup görüntü verilerini NumPy dizisine dönüştürür.
#--> bu çıktıyıda input değişkenine atıyoruz.
#---------------------------------------------------------------------------------------------------------------------

height, width = image.shape[:2]
#--> height ve width adında 2 tane değişken oluşturup bunu image ye atadığımız resmin yükseklik ve genişliğinin değerini atıyoruz.
#--> [:2] ifadesi, image.shape nın ilk iki öğesini alır, yani yükseklik ve genişlik.
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------

sobel_x = cv2.Sobel(image, cv2.CV_64F,0,1,ksize=5)
sobel_y = cv2.Sobel(image, cv2.CV_64F,1,0,ksize=5)
#--> cv2.Sobel() Bu fonksiyon bir görüntü üzerinde Sobel kenar tespitme filtresini uygular.
#--> Üç parametre alır. ilki filtrelencek resimimiz. ikincisi kenar tespitinde negatif değerlerin de alınabilmesi için 64 bit float tipi seçilmiştir. float tipini yazıyoruz.
#--> 0 ve 1 (veya 1 ve 0): X ve Y yönlü türevlerin sırasıyla uygulanacağı yönlendirme vektörleri. 0 X yönünü ve 1 Y yönünü temsil eder.
#--> Üçüncü parametre de Sobel çekirdeğinin boyutu. Burada, 5x5 boyutunda bir çekirdek kullanılır.
#---------------------------------------------------------------------------------------------------------------------

cv2.imshow("Rotated Image",image)
#--> cv2 nin imshow fonksiyonunu çağırıyoruz. bu fonksiyon okunan resim verilerini ekranda gösterir.
#--> fonksiyon içerine 2 parametre veriyoruz. ilk "Rotated Image" paremtresi başlık olarak veriyoruz. ikinci okunan veriyi ekranda göstermesi için atamış olduğumuz image parametresini veriyoruz
#--> çalıştırdığımızda küçük pencerede atadığımız resmi gösteriyor.
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------

cv2.waitKey(0)
#--> cv2 nin waitkey fonksiyonunu çağırıyoruz. bu fonksiyon herhangi bir tuşa basılmasını bekler.
#--> Bu fonksiyon, bir tuşa basılana kadar programın yürütülmesini durdurur.
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------

cv2.imshow("Sobel X",sobel_x)
#--> cv2 nin imshow fonksiyonunu çağırıyoruz. bu fonksiyon okunan resim verilerini ekranda gösterir.
#--> fonksiyon içerine 2 parametre veriyoruz. ilk "Sobel X" paremtresi başlık olarak veriyoruz. ikinci okunan veriyi ekranda göstermesi için atamış olduğumuz sobel_x parametresini veriyoruz
#--> çalıştırdığımızda küçük pencerede atadığımız resmi gösteriyor.
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------

cv2.waitKey(0)
#--> cv2 nin waitkey fonksiyonunu çağırıyoruz. bu fonksiyon herhangi bir tuşa basılmasını bekler.
#--> Bu fonksiyon, bir tuşa basılana kadar programın yürütülmesini durdurur.
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------

cv2.imshow("Sobel Y",sobel_y)
#--> cv2 nin imshow fonksiyonunu çağırıyoruz. bu fonksiyon okunan resim verilerini ekranda gösterir.
#--> fonksiyon içerine 2 parametre veriyoruz. ilk "Sobel Y" paremtresi başlık olarak veriyoruz. ikinci okunan veriyi ekranda göstermesi için atamış olduğumuz sobel_y parametresini veriyoruz
#--> çalıştırdığımızda küçük pencerede atadığımız resmi gösteriyor.
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------

cv2.waitKey(0)
#--> cv2 nin waitkey fonksiyonunu çağırıyoruz. bu fonksiyon herhangi bir tuşa basılmasını bekler.
#--> Bu fonksiyon, bir tuşa basılana kadar programın yürütülmesini durdurur.
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------

sobel_OR = cv2.bitwise_or(sobel_x,sobel_y)
#--> cv2 nin bitwise_or fonksiyonunu çağırıyoruz. Bu fonksiyon, iki giriş görüntüsü arasında bir veya (OR) işlemi gerçekleştirir.
#--> İki parametre alır. İlki X yönlü kenar tespitme operatörleri tarafından elde edilen görüntü.
#--> İkincisi de Y yönlü kenar tespitme operatörleri tarafından elde edilen görüntü.
#--> x ve y yönlü kenar tespitme operatörleri tarafından tespit edilen kenarlar arasında bir veya işlemi uygulanmış bir görüntü içerir.
#--> Bu, genellikle daha kapsamlı bir kenar tespiti sonucu elde etmek için kullanılır.

cv2.imshow("Sobel OR",sobel_OR)
#--> cv2 nin imshow fonksiyonunu çağırıyoruz. bu fonksiyon okunan resim verilerini ekranda gösterir.
#--> fonksiyon içerine 2 parametre veriyoruz. ilk "Sobel OR" paremtresi başlık olarak veriyoruz. ikinci okunan veriyi ekranda göstermesi için atamış olduğumuz sobel_OR parametresini veriyoruz
#--> çalıştırdığımızda küçük pencerede atadığımız resmi gösteriyor.
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------

cv2.waitKey(0)
#--> cv2 nin waitkey fonksiyonunu çağırıyoruz. bu fonksiyon herhangi bir tuşa basılmasını bekler.
#--> Bu fonksiyon, bir tuşa basılana kadar programın yürütülmesini durdurur.
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------

laplacian = cv2.Laplacian(image,cv2.CV_64F)
#--> cv2 nin Laplacian fonksiyonunu çağırıyoruz. Bu fonksiyon, kenar tespitme filtresini kullanarak bir görüntünün kenarlarını tespit eder.
#--> İki parametre alır. İlki resmimiz. İkincisi Türe dönüşüm tipi. Burada, kenar tespitinde negatif değerlerin de alınabilmesi için 64 bit float tipi seçilmiştir.
#--> laplacian değişkeni, görüntünün Laplacian kenar tespiti filtresi uygulanmış halini içerir.
#--> Bu, görüntünün genel kenar özelliklerini belirlemek için kullanılabilir.
#---------------------------------------------------------------------------------------------------------------------

cv2.imshow("Laplacian",laplacian)
#--> cv2 nin imshow fonksiyonunu çağırıyoruz. bu fonksiyon okunan resim verilerini ekranda gösterir.
#--> fonksiyon içerine 2 parametre veriyoruz. ilk "Laplacian" paremtresi başlık olarak veriyoruz. ikinci okunan veriyi ekranda göstermesi için atamış olduğumuz laplacian parametresini veriyoruz
#--> çalıştırdığımızda küçük pencerede atadığımız resmi gösteriyor.
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------

cv2.waitKey(0)
#--> cv2 nin waitkey fonksiyonunu çağırıyoruz. bu fonksiyon herhangi bir tuşa basılmasını bekler.
#--> Bu fonksiyon, bir tuşa basılana kadar programın yürütülmesini durdurur.
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------

canny = cv2.Canny(image,50,120)
#--> cv2 nin Canny fonksiyonunu çağırıyoruz. Bu fonksiyon, kenar tespitme yöntemini kullanarak bir görüntünün kenarlarını tespit eder.
#--> Üç parametre alır. ilki resmimiz. ikincisi ve üçüncüsü Eşik değerleri. Canny algoritması, kenarları tespit etmek için iki eşik değeri kullanır.
#--> İlk eşik değeri (50) düşük eşik, ikinci eşik değeri (120) ise yüksek eşiktir. Kenarlar, bu eşik değerleri arasında oluşturulan kenar pikselleri tarafından belirlenir.
#--> canny değişkeni, görüntünün Canny kenar tespiti algoritması uygulanmış halini içerir.
#--> Bu, genellikle diğer kenar tespitme yöntemlerine göre daha doğru sonuçlar sağlar ve kenarları belirlemek için yaygın olarak kullanılır.

cv2.imshow("Canny",canny)
#--> cv2 nin imshow fonksiyonunu çağırıyoruz. bu fonksiyon okunan resim verilerini ekranda gösterir.
#--> fonksiyon içerine 2 parametre veriyoruz. ilk "Canny" paremtresi başlık olarak veriyoruz. ikinci okunan veriyi ekranda göstermesi için atamış olduğumuz canny parametresini veriyoruz
#--> çalıştırdığımızda küçük pencerede atadığımız resmi gösteriyor.
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------

cv2.waitKey(0)
#--> cv2 nin waitkey fonksiyonunu çağırıyoruz. bu fonksiyon herhangi bir tuşa basılmasını bekler.
#--> Bu fonksiyon, bir tuşa basılana kadar programın yürütülmesini durdurur.
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------

cv2.destroyAllWindows()
#--> cv2 nin destroyAllWindows fonksiyonunu çağırıyoruz.bu fonksiyon opencv-python tarafından açılan pencereleri kapatır.
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
```

### Real Time Picture

```python
import cv2
#--> opencv-python paketinin Python'a aktarılan adıdır. import diyerek opencv-python kütüphanesini cv2 yazarak kullanıyoruz
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------

import numpy as np
#--> opencv-python paketinin numpy kütüphanesini import edip np numpy kütüphanesinin sıkça kullanılan bir kısaltmasıdır. 
#--> Numpy, Python programlama dilinde bilimsel hesaplamalar için kullanılan güçlü bir kütüphanedir.
#--> Numpy, çok boyutlu dizilerin (arrays) oluşturulması ve işlenmesi için çeşitli fonksiyonlar ve araçlar sağlar.
#--> Bu, matematiksel işlemler, lineer cebir, rastgele sayı üretimi, Fourier dönüşümleri gibi birçok alanı içerir.
#---------------------------------------------------------------------------------------------------------------------

def sketch(image):
    #--> sketch adında bir fonksiyon oluşturuyoruz. 
    #--> image adında bir parametre giriyoruz.
    #--> bu fonksiyonu çağırdığımızda image adında parametre almasını sağlıyoruz.
    #--> Bu fonksiyon görüntüyü alır ve bu görüntünün kenarlarını tespit eden basit bir eskiz oluşturur.
    #---------------------------------------------------------------------------------------------------------------------

    img_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    #--> cv2 nin cvtcolor fonksiyonunu çağırıyoruz. bu fonksiyon yüklediğimiz resimin renklerini dönüşmesini sağlar.
    #--> Fonksiyon 2 parametre alır. ilk renklerini değiştirceğimiz resim. ikincisi hangi renklerin neye değişeceği.
    #--> cv2 nin COLOR_BGR2GRAY BGR renk uzayından gri tonlamaya yapılacağını belirtir.
    #--> en sonda dönüştürülmüş gri ton resim verileri img_gray değişkenine atanır.
    #---------------------------------------------------------------------------------------------------------------------

    img_blur = cv2.GaussianBlur(img_gray,(5,5),0)
    #--> cv2 nin GaussianBlur fonksiyonunu çağırıyoruz. bu fonksiyon cvtcolor dönüşmesinden sonra Gri tonlamalı görüntüyü Gauss bulanıklığına dönüştürür.
    #--> Fonksiyon 3 parametre alır. (5, 5): Bulanıklık filtresinin boyutu. Bu, genellikle (tek ve pozitif) bir sayı veya bir tuple olarak belirtilir. Burada, 5x5 boyutunda bir filtre kullanılmıştır.
    #--> 0: X ve Y yönlerindeki standart sapma değeri. Bu, Gaussian fonksiyonunun şeklini belirler. 0 kullanıldığında, fonksiyon otomatik olarak standart sapmayı hesaplar.
    #--> en sonda dönüştürülmüş resmin verilerini img_blur değişkenine atanır.
    #---------------------------------------------------------------------------------------------------------------------

    img_canny = cv2.Canny(img_blur,10,70)
    #--> cv2 nin Canny fonksiyonunu çağırıyoruz. Gaussian bulanıklığı uygulandıktan sonra Canny kenar tespitme yöntemini kullanarak bir görüntünün kenarlarını tespit eder.
    #--> Üç parametre alır. ilki resmimiz. ikincisi ve üçüncüsü Eşik değerleri. Canny algoritması, kenarları tespit etmek için iki eşik değeri kullanır.
    #--> İlk eşik değeri (10) düşük eşik, ikinci eşik değeri (70) ise yüksek eşiktir. Kenarlar, bu eşik değerleri arasında oluşturulan kenar pikselleri tarafından belirlenir.
    #--> canny değişkeni, görüntünün Canny kenar tespiti algoritması uygulanmış halini içerir.
    #--> Bu, genellikle diğer kenar tespitme yöntemlerine göre daha doğru sonuçlar sağlar ve kenarları belirlemek için yaygın olarak kullanılır.

    return img_canny
    #--> Burda return yani döndür diyoruz.
    #--> Bu fonksiyonu çağırdığımızda ve image parametresini verdiğimizde resimde bir kaç işlem yaptıktan sonra en son halini döndürmesini sağlıyor.
    #--> Image parametresinde verilen resimi kenarlarını tespit edip basit bir eskiz oluşturduktan sonra o resimi döndürür.
    #--> Bu fonksiyonu değişkene atayabiliriz ve onu kullanabiliriz.
    #---------------------------------------------------------------------------------------------------------------------

cap = cv2.VideoCapture(0)
#--> cv2 nin VideoCapture fonksiyonunu çağırıyoruz. Bu kod, bilgisayarınıza bağlı bir kameradan video akışını yakalamak için bir VideoCapture nesnesi oluşturur.
#--> Bir parametre alır. Eğer parametre olarak bir dosya adı verilirse, o dosyadan video akışını okur.
#--> Eğer parametre olarak bir cihaz indeksi verilirse, bu, kullanılabilir kamera cihazlarının bir listesi olarak kabul edilir ve belirli bir kamerayı belirtmek için kullanılır. Örneğin, 0 ilk kamera cihazını belirtir.
#--> Bu görüntüyü de cap değişkenine atıyoruz.
#---------------------------------------------------------------------------------------------------------------------

while True:
    #--> While bir döngüdür.
    #--> verdiğimiz bool un doğru olduğu sürece çalışmasını sağlar.
    #--> Burda True diyerek sürekli çalışmasını sağlıyoruz
    #---------------------------------------------------------------------------------------------------------------------
    #---------------------------------------------------------------------------------------------------------------------

    ret,frame = cap.read()
    #--> Burda görüntüyü atadğımız değişken (cap) onun read fonksiyonunu çağırıyoruz. VideoCapture nesnesinden bir kareyi okur.
    #--> Her çağrıldığında, VideoCapture nesnesinin mevcut durumuna bağlı olarak, bir sonraki kareyi okur.
    #--> Bu yöntem iki değer döndürür: ret ve frame.
    #--> ret: Okuma işleminin başarılı olup olmadığını belirten bir boolean değerdir. Eğer okuma başarılıysa True, değilse False değeri döner.
    #--> frame: Okunan karenin görüntü verisini içeren bir NumPy dizisidir.

    cv2.imshow("Real Time Picture",sketch(frame))
    #--> cv2 nin imshow fonksiyonunu çağırıyoruz. bu fonksiyon okunan resim verilerini ekranda gösterir.
    #--> fonksiyon içerine 2 parametre veriyoruz. ilk "Real Time Picture" paremtresi başlık olarak veriyoruz. ikinci okunan veriyi ekranda göstermesi için atamış olduğumuz sketch(frame) parametresini veriyoruz
    #--> çalıştırdığımızda sketch fonksiyonunu çağırıp kameramızdan görüntü alacak.
    #--> Bu görüntüyü skecth fonksiyonu efekt ekleyip bize gösterecek.
    #---------------------------------------------------------------------------------------------------------------------

    if cv2.waitKey(1) == 13:
        break
    #--> if ile cv2 nin waitkey fonksiyonunu çağırıyoruz. parametre olarak 1 veriyoruz ve bu eşitse 13 e diyoruz.
    #--> Eğer 13 e eşitse if ten break diyoruz yani ifin içinden çıkıp devam et diyoruz.
    #--> waitkey fonksiyonu klavyeden olay bekler olayı alıp bir sonraki adıma geçmesini sağlar.
    #--> Parametre olarak milisaniye cinsinden bekleme süresi alır. 0 verilirse, sonsuza kadar bekler. Eğer bir tuşa basılırsa, bu tuşun ASCII kodunu döndürür. Eğer herhangi bir tuşa basılmazsa, -1 döner.
    #--> if == 13: Bu, dönen ASCII kodunu kontrol eder. Burada, Enter tuşunun ASCII kodu olan 13 ile karşılaştırılır.
    #--> kullanıcının Enter tuşuna basarak bir döngüyü sonlandırmasını sağlar.

cap.release()
#--> Burda görüntüyü atadğımız değişken (cap) onun release fonksiyonunu çağırıyoruz. VideoCapture nesnesinin kaynaklarını serbest bırakır.
#--> VideoCapture nesnesi tarafından kullanılan kamera veya video kaynağı gibi kaynakları serbest bırakır
#--> Bu kaynaklar artık kullanılmayacağı için serbest bırakılmaları önemlidir.
#--> Ayrıca, bu işlem, işletim sistemi kaynaklarının da geri alınmasını sağlar.
#--> programın çalışması sırasında kamera veya video kaynağına erişimi düzgün bir şekilde sonlandırmak için kullanılır.

cv2.destroyAllWindows()
#--> cv2 nin destroyAllWindows fonksiyonunu çağırıyoruz.bu fonksiyon opencv-python tarafından açılan pencereleri kapatır.
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
#---------------------------------------------------------------------------------------------------------------------
```
